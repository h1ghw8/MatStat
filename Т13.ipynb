```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

np.random.seed(123)

def s2(sig, k):
    return k / (k - 1) * sig**2

alp = 0.05

se_l = 5.722
se_w = 4.612
su_l = 6.161
su_w = 5.055

n1 = 139
n2 = 1000

v1 = s2(se_l, n1)
v2 = s2(se_w, n1)
v3 = s2(su_l, n2)
v4 = s2(su_w, n2)

f_l = v1 / v3
f_w = v2 / v4

ql = st.f.ppf(alp / 2, n1 - 1, n2 - 1)
qr = st.f.ppf(1 - alp / 2, n1 - 1, n2 - 1)

x = np.linspace(0.001, 5, 1000)
y = 1 - st.f.cdf(qr, n1 - 1, n2 - 1, scale=x) + st.f.cdf(ql, n1 - 1, n2 - 1, scale=x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='#2d6a4f', linewidth=2)
plt.axvline(x=1, color='#52b788', linestyle='--', linewidth=1.5)
plt.xlabel('θ')
plt.ylabel('W(θ)')
plt.title('Мощность критерия')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
