```python
import numpy as np
import scipy.stats as stats

# Параметры
n = 100
theta_true = 10
beta = 0.95
np.random.seed(46)

# Генерация выборки из равномерного распределения R[θ, 2θ]
sample = np.random.uniform(theta_true, 2 * theta_true, n)
x_max = np.max(sample)
x_mean = np.mean(sample)

# Оценки параметров
theta_omm = (2 / 3) * x_mean
theta_omp_corrected = ((n + 1) / (2 * n + 1)) * x_max

print(f"Выбранное theta: {theta_true}")
print(f"Выборка: {sample}")
print(f"Оценка метода моментов: {theta_omm:.4f}")
print(f"Оценка метода максимального правдоподобия (исправленная): {theta_omp_corrected:.4f}")
print()

# d) Точный доверительный интервал
t1 = 1 + ((1 - beta) / 2) ** (1 / n)
t2 = 1 + ((1 + beta) / 2) ** (1 / n)

exact_lower = x_max / t2
exact_upper = x_max / t1
exact_length = exact_upper - exact_lower

print("=" * 60)
print(" " * 20 + "Точный доверительный интервал")
print("=" * 60)
print(f"   ({exact_lower:.4f}, {exact_upper:.4f}), l = {exact_length:.4f}")
print("=" * 60)
print()

# e) Асимптотический доверительный интервал (на основе ОММ)
z1 = stats.norm.ppf((1 - beta) / 2)
z2 = stats.norm.ppf((1 + beta) / 2)
alpha1 = x_mean
alpha2 = 1 / n * np.sum(sample ** 2)
std_error = 2 / 3 / (n ** 0.5) * (alpha2 - alpha1 ** 2) ** 0.5

asympt_lower = theta_omm - std_error * z2
asympt_upper = theta_omm - std_error * z1
asympt_length = asympt_upper - asympt_lower

print("=" * 60)
print(" " * 15 + "Асимптотический доверительный интервал (ОММ)")
print("=" * 60)
print(f"   ({asympt_lower:.4f}, {asympt_upper:.4f}), l = {asympt_length:.4f}")
print("=" * 60)
print()

# g) Непараметрический bootstrap для ОММ
print("Непараметрический bootstrap для ОММ:")
bootstrap_iterations = 1000

bootstrap_delta_omm = []
for _ in range(bootstrap_iterations):
    boot_sample = np.random.choice(sample, size=n, replace=True)
    boot_theta = 2 / 3 * np.mean(boot_sample)
    bootstrap_delta_omm.append(boot_theta - theta_omm)

bootstrap_delta_omm_sorted = np.array(sorted(bootstrap_delta_omm))
alpha = 1 - beta
lower_percentile = int((alpha / 2) * bootstrap_iterations)
upper_percentile = int((1 - alpha / 2) * bootstrap_iterations)

boot_omm_lower = theta_omm - bootstrap_delta_omm_sorted[upper_percentile]
boot_omm_upper = theta_omm - bootstrap_delta_omm_sorted[lower_percentile]
boot_omm_length = boot_omm_upper - boot_omm_lower

print("=" * 70)
print(" " * 17 + "Непараметрический бутстраповский интервал (ОММ)")
print("=" * 70)
print(f"   ({boot_omm_lower:.4f}, {boot_omm_upper:.4f}), l = {boot_omm_length:.4f}")
print("=" * 70)
print()

# Непараметрический bootstrap для ОМП
print("Непараметрический bootstrap для ОМП:")
bootstrap_delta_omp = []
for _ in range(bootstrap_iterations):
    boot_sample = np.random.choice(sample, size=n, replace=True)
    boot_theta = (n + 1) / (2 * n + 1) * np.max(boot_sample)
    bootstrap_delta_omp.append(boot_theta - theta_omp_corrected)

bootstrap_delta_omp_sorted = np.array(sorted(bootstrap_delta_omp))

boot_omp_lower = theta_omp_corrected - bootstrap_delta_omp_sorted[upper_percentile]
boot_omp_upper = theta_omp_corrected - bootstrap_delta_omp_sorted[lower_percentile]
boot_omp_length = boot_omp_upper - boot_omp_lower

print("=" * 70)
print(" " * 17 + "Непараметрический бутстраповский интервал (ОМП)")
print("=" * 70)
print(f"   ({boot_omp_lower:.4f}, {boot_omp_upper:.4f}), l = {boot_omp_length:.4f}")
print("=" * 70)
print()

# h) Сравнение интервалов
intervals = [
    ("Точный", exact_length),
    ("Асимптотический (ОММ)", asympt_length),
    ("Непараметрический бутстрап (ОММ)", boot_omm_length),
    ("Непараметрический бутстрап (ОМП)", boot_omp_length)
]

sorted_intervals = sorted(intervals, key=lambda x: x[1])

print("=" * 60)
print(" " * 17 + "Доверительные интервалы (в порядке улучшения)")
print("=" * 60)
for name, length in sorted_intervals:
    print(f"   • {name}: l = {length:.4f}")
print("=" * 60)
```
