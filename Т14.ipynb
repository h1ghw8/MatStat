```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

np.random.seed(123)

x = np.array([-1.11, -6.10, 2.42])
y = np.array([-2.29, -2.91])
sgm_x_sq = 2
sgm_y_sq = 1
alpha = 0.05

mx = np.mean(x)
my = np.mean(y)
n = len(x)
m = len(y)

delta = (mx - my) / np.sqrt((sgm_x_sq/n + sgm_y_sq/m))
critic = st.norm.ppf(1-alpha)

print("Статистика критерия:", delta, "Критическая область G: delta >", critic)
print("Нет веских оснований отвергнуть H0" if delta < critic else "Отвергаем H0")

X = np.linspace(-1, 5, 1000)
quant = st.norm(0, np.sqrt(sgm_x_sq/n + sgm_y_sq/m)).ppf(1-alpha)
Y = 1 - st.norm(X, np.sqrt(sgm_x_sq/n + sgm_y_sq/m)).cdf(quant)

plt.figure(figsize=(10, 5))
plt.plot(X, Y, color='#2d6a4f', linewidth=2)
plt.axvline(x=0, color='#52b788', linestyle='--', linewidth=1.5)
plt.xlabel('θ')
plt.ylabel('W(θ)')
plt.title('Мощность критерия')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
